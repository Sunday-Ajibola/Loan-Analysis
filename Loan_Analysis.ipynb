###Outlier Detection

```python
#Boxplot to check for extreme values (outliers) in Monthly Debt.
sns.boxplot(x=credit["Monthly Debt"])
```

```python
#Calculating how many standard deviations each value in Monthly Debt is from the average.
from scipy.stats import zscore
credit["bad_debt"] = zscore(credit["Monthly Debt"])
credit
```

```python
#Identifying rows where the bad_debt is greater than 3 or less than -3 — which is considered an outlier.
outliers = credit[credit["bad_debt"].abs() > 3]
outliers
```

###Feature Engineering

```python
# Defining a custom function that categorizes credit scores into human-friendly labels.
def credit_category(score):
    if score >= 800:
        return 'Excellent'
    elif score >= 740:
        return 'Very Good'
    elif score >= 670:
        return 'Good'
    elif score >= 580:
        return 'Fair'
    elif score >= 300:
        return 'Poor'
    else:
        return 'Unknown'
```

```python
#Applying the function to each row and creates a new column called Credit_Category.
credit["Credit_Category"] = credit["Credit Score"].apply(credit_category)
credit
```

```python
# Creating a new column: Debt-to-Income ratio (DTI)
credit["DTI"] = credit["Monthly Debt"] / (credit["Annual Income"] / 12)
credit
```

###Group Comparisons

```python
# Grouping the data by Home Ownership type and calculating the average loan amount for each group.
print(credit.groupby("Home Ownership")["Current Loan Amount"].mean())
```
